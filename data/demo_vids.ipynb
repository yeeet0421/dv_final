```python
import json
## add vid sentiment
dict_ = {'世足賽': ['世足賽', '台灣'], '疫情': ['中心', '疫苗']}
f = open('demo_vidInfo.json')
info = json.load(f)
for keyword in dict_:
    f = open(keyword+'_vid_sentiment.json')
    sent = json.load(f)
    for word in dict_[keyword]:
        for vidInfo in info[keyword][word]:
            for vid in vidInfo:
                vidInfo[vid]['total_view'] = (sum(vidInfo[vid]['channel_chart']['viewcount']))
                vidInfo[vid]['total_video'] = (sum(vidInfo[vid]['channel_chart']['videocount']))
                vidInfo[vid]['avg_sentiment'] = sent[vid]['avg_sentiment']
with open('demo_vidInfo.json', 'w', encoding='utf-8') as f:
    json.dump(info, f, ensure_ascii=False, indent=4)
```

```python
import json
import math as math

## add com sentiment
dict_ = {'世足賽': ['世足賽', '台灣'], '疫情': ['中心', '疫苗']}
f = open('demo_vidInfo.json')
info = json.load(f)
for keyword in dict_:
    f = open(keyword+'_comment_sentiment.json')
    sent = json.load(f)
    for word in dict_[keyword]:
        for vidInfo in info[keyword][word]:
            for vid in vidInfo:
                bar = dict((i, 0) for i in (range(0, 10)))
                # print(vid, len((sent[vid])))
                # num = 0
                for com in sent[vid]:
                    # num += 1
                    # print(num)
                    if com["avg_sentiment"] == 1:
                        bar[9] += 1
                    else:
                        bar[math.floor(com["avg_sentiment"]*10%10)] += 1
                # print(bar)
                vidInfo[vid]['com_bar'] = list(bar.values())
                vidInfo[vid]['total_com'] = len(sent[vid])
                # vidInfo[vid]['avg_sentiment'] = sent[vid]['avg_sentiment']
with open('demo_vidInfo.json', 'w', encoding='utf-8') as f:
    json.dump(info, f, ensure_ascii=False, indent=4)
```

```python
import json

## overview bar chart
f = open('line.json')
view = json.load(f)

info = {'view':[], 'like':[], 'com':[], 'vid':[]}
for keyword in view:
    info['view'].append(view[keyword]['stat']['viewCount'])
    info['like'].append(view[keyword]['stat']['likeCount'])
    info['com'].append(view[keyword]['stat']['comCount'])
    info['vid'].append(view[keyword]['stat']['vidCount'])
with open('overview.json', 'w', encoding='utf-8') as f:
    json.dump(info, f, ensure_ascii=False, indent=4)
```
